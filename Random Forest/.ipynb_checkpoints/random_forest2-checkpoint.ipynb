# Parameters
- n_estimators: Số lượng cây quyết định trong rừng.
- criterion: Hàm loss để đánh giá chất lượng của một split.
- max_depth: Độ sâu tối đa của cây. Nếu không chỉ định, các cây sẽ được mở rộng cho đến khi tất cả các lá chỉ chứa một mẫu hoặc không thể tách thêm được nữa.
- min_samples_split: Số lượng mẫu tối thiểu cần thiết để một nút có thể tách.
- min_samples_leaf: Số lượng mẫu tối thiểu cần thiết để một nút lá có thể tồn tại.
- min_weight_fraction_leaf: Phần trọng số tối thiểu cần thiết để một nút lá có thể tồn tại.
- max_features: Số lượng features tối đa được xem xét khi tìm kiếm split tốt nhất.
- max_leaf_nodes: Số lượng nút lá tối đa trong cây.
- min_impurity_decrease: Một split sẽ chỉ được thực hiện nếu nó làm giảm impurity ít nhất một giá trị nào đó.
- bootstrap: Cho phép hoặc không cho phép việc sử dụng bootstrap samples khi xây dựng cây.
- oob_score: Cho phép hoặc không cho phép việc sử dụng các mẫu ngoài bag để đánh giá độ chính xác tổng quát.
- n_jobs: Số lượng job được chạy song song. -1 có nghĩa là sử dụng tất cả các CPU.
- random_state: Seed ngẫu nhiên để đảm bảo tính tái lập.
- verbose: Điều khiển mức độ in log.
- warm_start: Khi được đặt thành True, re-use the solution of the previous call to fit and add more estimators to the ensemble.
- ccp_alpha: Giá trị phạt độ phức tạp sử dụng để pruning cây quyết định.
- max_samples: Số lượng mẫu tối đa được vẽ từ dữ liệu gốc để đào tạo mỗi cây. Nếu None, tất cả các mẫu sẽ được sử dụng.
